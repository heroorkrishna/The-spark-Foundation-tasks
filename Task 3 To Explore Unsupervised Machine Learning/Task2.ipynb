```python
import tensorflow as tf
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
seed = 1234
np.random.seed(seed)
tf.set_random_seed(seed)
```

```python
dataset = pd.read_csv('Iris_Dataset.csv')
```

```python
dataset.head()
```

```python
g=sns.pairplot(dataset,size = 2.5)
```

```python
cols = dataset.columns
features = cols[0:4]
labels = cols[4]
print(features)
print(labels)
```

```python
data_norm = pd.DataFrame(dataset)

for feature in features:
    dataset[feature] = (dataset[feature] - dataset[feature].mean())/dataset[feature].std()
```

```python
print("Averages")
print(dataset.mean())

print("\n Deviations")
print(pow(dataset.std(),2))
```

```python
#Shuffle The data
indices = data_norm.index.tolist()
indices = np.array(indices)
np.random.shuffle(indices)
X = data_norm.reindex(indices)[features]
y = data_norm.reindex(indices)[labels]
```

```python
dataset = pd.get_dummies(dataset, columns=['Species']) #One Hot Encoding
values = list(dataset.columns.values)
```

```python
y = dataset[values[-3:]]
y = np.array(y, dtype='float32')
X = dataset[values[1:-3]]
X = np.array(X, dtype='float32')
```

```python
# Shuffle Data
indices = np.random.choice(len(X), len(X), replace=False)
X_values = X[indices]
y_values = y[indices]
```

```python
# Creating a Train and a Test Dataset
test_size = 10
X_test = X_values[-test_size:]
X_train = X_values[:-test_size]
y_test = y_values[-test_size:]
y_train = y_values[:-test_size]
```

```python
sess = tf.Session()
interval = 25
epoch = 500
```

```python
X_data = tf.placeholder(shape=[None, 4], dtype=tf.float32)
y_target = tf.placeholder(shape=[None, 3], dtype=tf.float32)
```

```python
hidden_layer_nodes = 8
```

```python
# nn layer
w1 = tf.Variable(tf.random_normal(shape=[4,hidden_layer_nodes])) 
b1 = tf.Variable(tf.random_normal(shape=[hidden_layer_nodes]))   
w2 = tf.Variable(tf.random_normal(shape=[hidden_layer_nodes,3])) 
b2 = tf.Variable(tf.random_normal(shape=[3]))
```

```python
hidden_output = tf.nn.relu(tf.add(tf.matmul(X_data, w1), b1))
final_output = tf.nn.softmax(tf.add(tf.matmul(hidden_output, w2), b2))
```

```python
loss = tf.reduce_mean(-tf.reduce_sum(y_target * tf.log(final_output), axis=0))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.001).minimize(loss)
```

```python
init = tf.global_variables_initializer()
sess.run(init)
```

```python
# Training
ls = {}
print('Training the model...')
for i in range(1, (epoch + 1)):
    sess.run(optimizer, feed_dict={X_data: X_train, y_target: y_train})
    if i % interval == 0:
        l = sess.run(loss, feed_dict={X_data: X_train, y_target: y_train})
        print('Epoch', i, '|', 'Loss:', l)
        ls[i] = l
```

```python
import matplotlib.pyplot as plt
iterations = list(ls.keys())
costs = list(ls.values())
plt.plot(iterations,costs)  
plt.ylabel('cost')
plt.xlabel('iterations (per fifties)')
plt.title("Learning rate = " + str(.001)) 
plt.show() 
```
